```python
import numpy as np
from tqdm import tqdm
import time
import json
import os
from datetime import datetime
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
import warnings
warnings.filterwarnings('ignore')

class BinaryGSAVariant:
    def __init__(self, variant_params, pop_size=30, max_iter=1000):
        self.pop_size = pop_size
        self.max_iter = max_iter
        self.G0 = variant_params['G0']
        self.alpha = variant_params['alpha']
        self.w_type = variant_params.get('w_type', 'static')
        self.w = variant_params.get('w', 0.7)
        self.w_start = variant_params.get('w_start', 0.9)
        self.w_end = variant_params.get('w_end', 0.4)
        self.force_limit = variant_params.get('force_limit', 100)
        self.convergence_curve = []
        
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-np.clip(x, -10, 10)))
    
    @staticmethod
    def evaluate_population(positions, func):
        fitness_values = np.zeros(len(positions))
        for i in range(len(positions)):
            try:
                raw_fitness = -float(func(positions[i]))
                fitness_values[i] = raw_fitness if raw_fitness > -1e10 else 0
            except:
                continue
        return fitness_values

    def repair_solution(self, solution, func):
        if -float(func(solution)) > -1e10:
            return solution
        
        selected = np.where(solution == 1)[0]
        if len(selected) <= 1:
            return np.zeros_like(solution)
        
        repaired = solution.copy()
        np.random.shuffle(selected)
        
        for idx in selected:
            repaired[idx] = 0
            if -float(func(repaired)) > -1e10:
                break
        
        return repaired

    def calculate_forces(self, positions, masses, G, k_best):
        n_particles, dimension = positions.shape
        forces = np.zeros((n_particles, dimension))
        best_indices = np.argsort(masses)[-k_best:]
        
        for i in range(n_particles):
            total_force = np.zeros(dimension)
            for j in best_indices:
                if i != j:
                    diff = positions[j] - positions[i]
                    R = np.sum(diff ** 2)
                    R = max(R, 1e-10)
                    force_magnitude = G * masses[i] * masses[j] / np.sqrt(R)
                    force_magnitude = np.clip(force_magnitude, -self.force_limit, self.force_limit)
                    total_force += force_magnitude * diff
            forces[i] = total_force
        return forces

   
```

```python
def optimize(self, objective_func, dimension):
        positions = (np.random.random((self.pop_size, dimension)) < 0.3).astype(np.int8)
        velocities = np.zeros((self.pop_size, dimension))
        
        # Pre-allocate random arrays for the entire run
        all_r1 = np.random.random((self.max_iter, self.pop_size, dimension))
        all_r2 = np.random.random((self.max_iter, self.pop_size, dimension))
        pos_rand = np.random.random((self.max_iter, self.pop_size, dimension))
        
        fitness = self.evaluate_population(positions, objective_func)
        best_fitness = np.max(fitness)
        best_position = positions[np.argmax(fitness)].copy()
        
        self.convergence_curve = [best_fitness]
        
        for t in range(self.max_iter):
            G = self.G0 * np.exp(-self.alpha * t / self.max_iter)
            
            fitness = self.evaluate_population(positions, objective_func)
            current_best = np.argmax(fitness)
            if fitness[current_best] > best_fitness:
                best_fitness = fitness[current_best]
                best_position = positions[current_best].copy()
            
            self.convergence_curve.append(best_fitness)
            
            fitness_range = np.max(fitness) - np.min(fitness)
            if fitness_range > 0:
                masses = (fitness - np.min(fitness)) / fitness_range
            else:
                masses = np.ones(self.pop_size) / self.pop_size
            masses = masses / (np.sum(masses) + 1e-10)
            
            k_best = max(2, int(self.pop_size * (1 - t/self.max_iter)))
            forces = self.calculate_forces(positions, masses, G, k_best)
            
            if self.w_type == 'dynamic':
                w = self.w_start - (self.w_start - self.w_end) * (t / self.max_iter)
            else:
                w = self.w
                
            velocities = w * velocities + forces
            velocities = np.clip(velocities, -4, 4)
            
            probabilities = self.sigmoid(velocities)
            positions = (pos_rand[t] < probabilities).astype(np.int8)
            
            for i in range(self.pop_size):
                positions[i] = self.repair_solution(positions[i], objective_func)
        
        return best_position, best_fitness, self.convergence_curve
```

```python
def create_convergence_plots(problems, convergence_data, variants, save_dir, include_variance=True):
    fig, axes = plt.subplots(2, 5, figsize=(25, 12))
    axes = axes.flatten()
    colors = ['blue', 'red', 'green', 'purple']
    
    for idx, (problem_name, (func, dim)) in enumerate(problems.items()):
        ax = axes[idx]
        
        for i, variant in enumerate(variants):
            histories = np.array(convergence_data[problem_name][variant['name']])
            mean_curve = np.mean(histories, axis=0)
            
            if variant['w_type'] == 'dynamic':
                param_str = f"{variant['name']}\n(G0={variant['G0']}, α={variant['alpha']}\nw={variant['w_start']}→{variant['w_end']})"
            else:
                param_str = f"{variant['name']}\n(G0={variant['G0']}, α={variant['alpha']}\nw={variant['w']})"
            
            ax.plot(mean_curve, 
                   color=colors[i],
                   label=param_str if idx == 0 else "", 
                   linewidth=2.5)
            
            if include_variance:
                std_curve = np.std(histories, axis=0)
                ax.fill_between(range(len(mean_curve)),
                              mean_curve - std_curve,
                              mean_curve + std_curve,
                              color=colors[i],
                              alpha=0.15)
        
        ax.set_title(f'{problem_name} Convergence\n(dim={dim})', 
                    fontsize=11, pad=10)
        ax.set_xlabel('Iteration', fontsize=10)
        ax.set_ylabel('Profit', fontsize=10)
        ax.grid(True, linestyle='--', alpha=0.5)
        ax.tick_params(axis='both', which='major', labelsize=9)
        
        if ax.get_ylim()[1] > 1e5:
            formatter = ScalarFormatter(useMathText=True)
            formatter.set_scientific(True)
            formatter.set_powerlimits((-2, 2))
            ax.yaxis.set_major_formatter(formatter)
    
    legend = fig.legend(bbox_to_anchor=(1.02, 0.5), loc='center left', 
                       fontsize=10, frameon=True, fancybox=True, shadow=True)
    legend.get_frame().set_alpha(0.9)
    
    plt.tight_layout()
    
    filename = 'convergence_with_variance.png' if include_variance else 'convergence_mean_only.png'
    plt.savefig(os.path.join(save_dir, filename), dpi=300, bbox_inches='tight')
    plt.close()

def run_comparative_experiments(mkp_problems, variants, n_runs=3):
    results = {v['name']: {} for v in variants}
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    save_dir = f"bgsa_comparison_results_{timestamp}"
    os.makedirs(save_dir, exist_ok=True)
    convergence_data = {prob: {v['name']: [] for v in variants} for prob in mkp_problems.keys()}

    for problem_name, (func, dimension) in mkp_problems.items():
        print(f"\nSolving {problem_name}")
        
        for variant in variants:
            variant_results = []
            print(f"\nTesting {variant['name']}")
            
            for run in tqdm(range(n_runs), desc=f"Runs for {problem_name} - {variant['name']}"):
                try:
                    optimizer = BinaryGSAVariant(variant)
                    best_sol, best_fit, conv = optimizer.optimize(func, dimension)
                    variant_results.append(float(best_fit))
                    convergence_data[problem_name][variant['name']].append(conv)
                except Exception as e:
                    print(f"Error in {variant['name']}, run {run}: {str(e)}")
            
            if variant_results:
                results[variant['name']][problem_name] = {
                    'Best': float(max(variant_results)),
                    'Mean': float(np.mean(variant_results)),
                    'StdDev': float(np.std(variant_results))
                }

    # Create plots
    create_convergence_plots(mkp_problems, convergence_data, variants, save_dir)
    create_convergence_plots(mkp_problems, convergence_data, variants, save_dir, include_variance=False)
    
    # Save results
    with open(os.path.join(save_dir, 'comparison_results.json'), 'w') as f:
        json.dump(results, f, indent=4)
    
    return results, save_dir
```
