# Algorithme d'Optimisation des Baleines Binaire Amélioré
---

## Caractéristiques WOA Original
- **Encerclement de la Proie**: Baleines se déplaçant vers la meilleure solution
- **Attaque en Spirale**: Mouvement en spirale vers la cible
- **Recherche Aléatoire**: Exploration de l'espace de recherche

## Nos Améliorations

### 1. Ajout du Croisement
Croisement à deux points entre solutions avec 40% de probabilité pour combiner les bonnes parties des solutions.

### 2. Ajout de la Recherche Locale
Mécanisme simple d'inversion de bits pour améliorer localement les solutions.

### 3. Ajout de l'Historique des Succès
Suivi des améliorations:
- Succès → augmentation du score (+0.1)
- Échec → diminution du score (-0.05)

### 4. Ajout du Momentum
Lisse le mouvement des baleines entre les itérations pour une meilleure convergence.

### 5. Paramètres Rendus Adaptatifs
- Rayon de recherche s'ajuste avec la progression
- Probabilité de stratégie basée sur le succès
- Meilleur équilibre exploration/exploitation

### 6. Ajout d'Amélioration Régulière
Recherche locale sur la meilleure solution tous les 10 itérations.

## Avantages
- Meilleure exploitation locale
- Diversité maintenue
- Calcul efficace
- Convergence plus fluide
- Multiples stratégies d'amélioration

Ces ajouts aident l'algorithme à mieux gérer le problème du sac à dos tout en conservant l'inspiration de la chasse des baleines.

```python
class BWOA:
    """
    Binary Whale Optimization Algorithm for Multiple Knapsack Problem
    Implements hunting behavior of humpback whales with binary adaptation
    """
    def __init__(self, pop_size=30, max_iter=1000):
        # Population parameters
        self.pop_size = pop_size        # Number of whales in population
        self.max_iter = max_iter        # Maximum iterations
        
        # Whale movement parameters
        self.a_start = 2.5             # Initial search radius
        self.a_end = 0.1              # Final search radius
        self.b = 1.5                  # Spiral shape constant
        self.p_crossover = 0.4        # Crossover probability
        self.local_search_prob = 0.1  # Probability of local search
        
        # Enhanced algorithm features
        self.adaptive_a = True        # Use adaptive search radius
        self.adaptive_p = True        # Adapt hunting strategy probability
        self.momentum = 0.1           # Momentum coefficient for position updates
        
        # Arrays for population management
        self.random_numbers = None    # Pre-allocated random numbers
        self.positions = None         # Whale positions (binary solutions)
        self.velocities = None        # Whale velocities
        self.fitness = None           # Solution fitness values

    def sigmoid(self, x):
        """Convert continuous values to probability in [0,1] range"""
        return 1 / (1 + np.exp(-np.clip(x, -10, 10)))
    
    def evaluate_batch(self, positions, func):
        """
        Evaluate fitness for multiple solutions efficiently
        Returns fitness values with penalty for infeasible solutions
        """
        fitness = np.zeros(len(positions))
        for i, pos in enumerate(positions):
            value = -func(pos)  # Negate because we're minimizing
            fitness[i] = value if value > -1e10 else 0
        return fitness
    
    def repair_solution(self, solution, func):
        """
        Repair infeasible solutions by removing items until constraints are satisfied
        Args:
            solution: Binary solution vector
            func: Objective function with constraints
        Returns:
            Feasible solution
        """
        if -func(solution) > -1e10:  # Solution is feasible
            return solution
            
        repaired = solution.copy()
        selected = np.where(repaired == 1)[0]  # Get indices of selected items
        
        if len(selected) <= 1:  # Can't repair if only one item
            return np.zeros_like(solution)
                
        np.random.shuffle(selected)  # Randomize removal order
        
        # Remove items until feasible
        for idx in selected:
            repaired[idx] = 0
            if -func(repaired) > -1e10:
                break
        
        return repaired
    
    def crossover(self, parent1, parent2):
        """
        Two-point crossover between solutions
        Args:
            parent1, parent2: Binary solution vectors
        Returns:
            Child solution
        """
        n = len(parent1)
        points = np.sort(np.random.choice(n, 2, replace=False))
        child = parent1.copy()
        child[points[0]:points[1]] = parent2[points[0]:points[1]]
        return child
    
    def local_search(self, solution, func, max_flips=5):
        """
        Simple local search by bit flipping
        Args:
            solution: Binary solution vector
            func: Objective function
            max_flips: Maximum bits to flip
        Returns:
            Improved solution
        """
        improved = solution.copy()
        current_fitness = -func(improved)
        
        # Try flipping a limited number of bits
        indices = np.random.choice(len(solution), min(max_flips, len(solution)), replace=False)
        for i in indices:
            neighbor = improved.copy()
            neighbor[i] = 1 - neighbor[i]  # Flip bit
            neighbor = self.repair_solution(neighbor, func)
            neighbor_fitness = -func(neighbor)
            
            if neighbor_fitness > current_fitness:
                improved = neighbor
                current_fitness = neighbor_fitness
                
        return improved
```

```python
def optimize(self, func, dimension):
        """
        Main optimization loop
        Args:
            func: Objective function (knapsack with constraints)
            dimension: Problem dimension (number of items)
        Returns:
            best_position: Best solution found
            best_fitness: Best fitness value
            convergence_history: Fitness history over iterations
        """
        # Initialize population with low density (25% ones) for knapsack
        self.positions = np.random.binomial(1, 0.25, (self.pop_size, dimension))
        self.velocities = np.zeros((self.pop_size, dimension))
        
        # Pre-allocate random numbers for efficiency
        self.random_numbers = np.random.random((self.max_iter, self.pop_size, 4))
        
        # Repair initial solutions
        for i in range(self.pop_size):
            self.positions[i] = self.repair_solution(self.positions[i], func)
        
        # Initial population evaluation
        self.fitness = self.evaluate_batch(self.positions, func)
        best_idx = np.argmax(self.fitness)
        best_position = self.positions[best_idx].copy()
        best_fitness = self.fitness[best_idx]
        
        # History tracking
        success_history = np.zeros(self.pop_size)  # Track individual whale success
        convergence_history = [best_fitness]
        
        # Main optimization loop
        for t in range(self.max_iter):
            # Update search radius 'a' - controls exploration vs exploitation
            if self.adaptive_a:
                progress = t / self.max_iter
                a = self.a_start - (self.a_start - self.a_end) * (1 - np.exp(-5 * progress))
            else:
                a = self.a_start - (self.a_start - self.a_end) * (t / self.max_iter)
            
            # Use pre-generated random numbers
            r1, r2 = self.random_numbers[t, :, :2].T
            
            # Update each whale's position
            for i in range(self.pop_size):
                # Attempt local search with probability
                if self.random_numbers[t, i, 2] < self.local_search_prob:
                    self.positions[i] = self.local_search(self.positions[i], func)
                    continue
                
                # Calculate search coefficients
                A = 2 * a * r1[i] - a  # Controls step size
                C = 2 * r2[i]          # Random coefficient
                
                # Determine hunting strategy probability
                p = 0.5 * (1 + success_history[i]) if self.adaptive_p else 0.5
                
                # Choose hunting behavior
                if p < 0.5:  # Encircling prey or searching
                    if abs(A) < 1:  # Encircling prey (exploitation)
                        D = np.abs(C * best_position - self.positions[i])
                        new_position = best_position - A * D
                    else:  # Searching for prey (exploration)
                        random_whale = self.positions[np.random.randint(self.pop_size)]
                        D = np.abs(C * random_whale - self.positions[i])
                        new_position = random_whale - A * D
                else:  # Spiral hunting pattern
                    l = self.random_numbers[t, i, 3] * 2 - 1
                    D = np.abs(best_position - self.positions[i])
                    spiral_mod = np.exp(self.b * l) * np.cos(2 * np.pi * l)
                    new_position = D * spiral_mod + best_position
                
                # Update velocity and position with momentum
                self.velocities[i] = (self.momentum * self.velocities[i] + 
                                    (1 - self.momentum) * new_position)
                # Convert to binary using sigmoid
                self.positions[i] = (np.random.random(dimension) < 
                                   self.sigmoid(self.velocities[i])).astype(int)
                
                # Selective crossover with best solution
                if np.random.random() < self.p_crossover:
                    self.positions[i] = self.crossover(self.positions[i], best_position)
                # Repair if constraints violated
                self.positions[i] = self.repair_solution(self.positions[i], func)
                
                # Update fitness and success history
                new_fitness = -func(self.positions[i])
                if new_fitness > self.fitness[i]:
                    success_history[i] = min(1.0, success_history[i] + 0.1)
                else:
                    success_history[i] = max(0.0, success_history[i] - 0.05)
                self.fitness[i] = new_fitness
            
            # Update best solution
            current_best = np.argmax(self.fitness)
            if self.fitness[current_best] > best_fitness:
                best_position = self.positions[current_best].copy()
                best_fitness = self.fitness[current_best]
            
            # Periodic local search on best solution
            if t % 10 == 0 and t > 0:
                improved_best = self.local_search(best_position, func)
                improved_fitness = -func(improved_best)
                if improved_fitness > best_fitness:
                    best_position = improved_best
                    best_fitness = improved_fitness
            
            convergence_history.append(best_fitness)
        
        return best_position, best_fitness, convergence_history
```

```python
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
from datetime import datetime
import json
import os
import time


def run_experiments(problems, n_runs=30):
    results = {}
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    save_dir = f"BWOA{timestamp}"
    os.makedirs(save_dir, exist_ok=True)
    
    convergence_data = {prob: [] for prob in problems}
    
    for problem_name, (func, dimension) in problems.items():
        print(f"\nSolving {problem_name}")
        problem_results = []
        histories = []
        
        for run in tqdm(range(n_runs)):
            optimizer = BWOA(pop_size=30, max_iter=1000)   
            best_solution, best_fitness, history = optimizer.optimize(func, dimension)
            problem_results.append(best_fitness)
            histories.append(history)
            
        results[problem_name] = {
            'Best': float(max(problem_results)),
            'Mean': float(np.mean(problem_results)),
            'StdDev': float(np.std(problem_results))
        }
        convergence_data[problem_name] = histories
    
    # Save results
    with open(os.path.join(save_dir, 'BWOA.json'), 'w') as f:
        json.dump(results, f, indent=4)
        
    return results, save_dir, convergence_data

def create_convergence_plots(problems, convergence_data, save_dir):
    fig, axes = plt.subplots(2, 5, figsize=(25, 12))
    axes = axes.flatten()
    
    for idx, (problem_name, (func, dim)) in enumerate(problems.items()):
        ax = axes[idx]
        histories = np.array(convergence_data[problem_name])
        
        if len(histories) > 0:
            # Ensure all sequences have same length
            min_len = min(len(h) for h in histories)
            histories = np.array([h[:min_len] for h in histories])
            
            mean_curve = np.mean(histories, axis=0)
            std_curve = np.std(histories, axis=0)
            
            ax.plot(mean_curve, color='blue', label='Mean', linewidth=2.5)
            ax.fill_between(range(len(mean_curve)),
                          mean_curve - std_curve,
                          mean_curve + std_curve,
                          color='blue', alpha=0.15)
            
        ax.set_title(f'{problem_name} Convergence\n(dim={dim})', fontsize=11, pad=10)
        ax.set_xlabel('Iteration', fontsize=10)
        ax.set_ylabel('Profit', fontsize=10)
        ax.grid(True, linestyle='--', alpha=0.5)
        
        if ax.get_ylim()[1] > 1e5:
            ax.yaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
    
    plt.tight_layout()
    plt.savefig(os.path.join(save_dir, 'convergence_plots.png'), dpi=300, bbox_inches='tight')
    plt.close()
```
